```python
import pandas as pd
```

```python
circuits_df = pd.read_csv('dataset/circuits.csv')
circuits_df.head()
```

```python
races_df = pd.read_csv('dataset/races.csv')
races_df.head()
```

```python
results_df = pd.read_csv('dataset/results.csv')
results_df.head()
```

```python
status_df = pd.read_csv('dataset/status.csv')
status_df.head()
```

```python
cir_races_df = pd.merge(circuits_df, races_df, on = 'circuitId')
cir_races_df.head()
```

```python
final_df = pd.merge(cir_races_df, results_df, on = 'raceId')
final_df.head()
```

```python
df_cir_status = pd.merge(final_df, status_df, on = 'statusId')
df_cir_status.head()
```

```python
finished_df_14 = df_cir_status[df_cir_status.circuitId == 14]
finished_df_14 = finished_df_14[finished_df_14.status == 'Finished']
finished_df_14
```

```python
df = final_df.dropna()
df.head()
```

```python
df[df.eq("N").any(1)]
```

```python
df_final = df[df.fastestLapSpeed != "\\N"]
df_final.head()
```

```python
df_final = df_final[df_final.alt != "\\N"]
df_final.head()
```

```python
df_alt_sorted = df_final.sort_values('alt')
df_alt_sorted.head()
```

```python
df_ = df_final.groupby(['circuitId'], group_keys=True).apply(lambda x: x)
df_.head()
```

```python
df_fastest = df_final.groupby(['circuitId', 'alt'])['fastestLapSpeed'].max()
df_fastest.head()
```

```python
df_fastest.to_csv('dataset/alt_fastest_speed.csv')
```
